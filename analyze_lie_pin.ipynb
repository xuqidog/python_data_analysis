# 招聘薪资统计

```python
# 招聘薪资
money_count = all_data['compensation'].value_counts()[:20]
money_count
```

```python
# 格式化薪资
import re
# 去除标签
def get_money_low(tag): 
    if tag == '面议':
        return '面议'
    else:
        items = tag.strip().split('-')
        return items[0]

def get_money_heigh(tag): 
    if tag == '面议':
        return '面议'
    else:
        items = tag.strip().split('-')
        return items[1].split('万')[0]
def get_money_mean(tag):
    if tag == '面议':
        return '面议'
    else:
        items = tag.strip().split('-')
        low = int(items[0])
        heigh = int(items[1].split('万')[0])
        return (low+heigh)/2
    
all_data['low'] = all_data['compensation'].apply(get_money_low)
all_data['heigh'] = all_data['compensation'].apply(get_money_heigh)
all_data['mean_money'] = all_data['compensation'].apply(get_money_mean)
all_data.to_csv('all_data.csv')
# all_data.head()
```

```python
from pylab import *
n = 12
X = np.arange(n)
Y1 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)
Y2 = (1-X/float(n)) * np.random.uniform(0.5,1.0,n)

bar(X, +Y1, facecolor='#9999ff', edgecolor='white')
bar(X, -Y2, facecolor='#ff9999', edgecolor='white')

for x,y,y2 in zip(X,Y1,Y2):
    text(x, y+0.05, '%.2f' % y, ha='center', va= 'bottom')
    text(x, -y2-0.15, '%.2f' % y2, ha='center', va= 'bottom')

ylim(-1.25,+1.25)
show()
```
